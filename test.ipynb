```python
from pyspark.sql import SparkSession

# 1. Initialize Spark
spark = SparkSession.builder.appName("investigation").master("local[*]").getOrCreate()

# 2. Read the parquet file
df = spark.read.parquet("o.parquet") # Assuming o.parquet is in your /app folder

# 3. Replicate the exact sampling logic from prepare_data.py
n = 1200
fraction = n / df.count()
sampled_df = df.select(["id", "title", "text"]).sample(fraction=fraction, seed=0).limit(n)

sampled_count = sampled_df.count()
print(f"1. Rows generated by Spark .sample(): {sampled_count}")
```
